1. The dataset contains information about movies and TV shows available on Netflix.
2. It includes attributes such as title, director, cast, country, release year, rating, and genre.
3. Both movies and TV shows are present in the dataset.
4. Some columns contain missing values, especially director, cast, country, and date_added.
5. The dataset can be used to analyze content trends, ratings, countries, and genre distribution.

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['director'].fillna('Not Available', inplace=True)
```

Missing values in the director column were replaced with 'Not Available' so that rows are retained for analysis.

```python
df['cast'].fillna('Not Available', inplace=True)
```

Missing values in the cast column were replaced with 'Not Available' to preserve all records.

```python
df['country'].fillna('Unknown', inplace=True)
```

Missing country values were replaced with 'Unknown' to avoid dropping useful data.

```python
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

The date_added column was converted to datetime format to enable time-based analysis.

```python
df['year_added'] = df['date_added'].dt.year
```

```python
df['month_added'] = df['date_added'].dt.month
```

```python
df.head()
```

Two new columns, year_added and month_added, were created to analyze trends in content additions over time.

```python
df.isnull().sum()
```

```python
#Step 3
#Exploratory Data Analysis (5 Questions)

#Question 1: How many Movies and TV Shows are there?

df['type'].value_counts()
```

This analysis shows the distribution of Movies and TV Shows available on Netflix.

```python
#Question 2: Which are the Top 10 countries producing Netflix content?

df['country'].value_counts().head(10)
```

This analysis identifies the countries contributing the most content to Netflix.

```python
#Question 3: Which content ratings are most common?

df['rating'].value_counts()
```

This analysis helps understand the audience categories targeted by Netflix content.

```python
#Question 4: How has the number of titles released changed over the years?

df['release_year'].value_counts().sort_index()
```

This analysis shows the trend in content releases across different years.

```python
#Question 5: Which genres are most popular?

df['listed_in'].value_counts().head(10)
```

This analysis reveals the most frequently occurring genres on Netflix.

```python
#Number of Movies and TV Shows added each year

df.groupby('year_added')['type'].count()
```

```python
#Average release year by type

df.groupby('type')['release_year'].mean()
```

```python
#Step 4
#Visualization
#1. Bar Chart
#Question: What is the distribution of Movies and TV Shows?

plt.figure(figsize=(6,4))
sns.countplot(x='type', data=df)

plt.title('Distribution of Movies and TV Shows')
plt.xlabel('Content Type')
plt.ylabel('Count')

plt.show()
```

```python
#2. Line Chart
#Question: How has the number of titles added to Netflix changed over the years?

year_data = df['year_added'].value_counts().sort_index()

plt.figure(figsize=(10,5))
plt.plot(year_data.index, year_data.values)

plt.title('Content Added to Netflix Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles')

plt.show()
```

```python
#3. Histogram
#Question: How are release years distributed?

plt.figure(figsize=(10,5))
plt.hist(df['release_year'], bins=20)

plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Frequency')

plt.show()
```

```python
#4. Scatter Plot
#Question: What is the relationship between release year and year added?

plt.figure(figsize=(8,6))

plt.scatter(df['release_year'], df['year_added'])

plt.title('Release Year vs Year Added')
plt.xlabel('Release Year')
plt.ylabel('Year Added')

plt.show()
```

```python
#5. Pie Chart
#Question: What percentage of Netflix content consists of Movies and TV Shows?

df['type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)

plt.title('Percentage of Movies and TV Shows')
plt.ylabel('')

plt.show()
```

```python
#6. Heatmap
#First create a numerical dataframe:

numeric_df = df[['release_year', 'year_added', 'month_added']]

plt.figure(figsize=(8,5))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```

Step 5: Insights Report

1. Movies significantly outnumber TV Shows on Netflix, indicating that the platform focuses more on movie content.

2. The United States contributes the highest number of titles, making it the dominant content producer on Netflix.

3. The number of titles added to Netflix increased rapidly after 2015, showing a period of strong expansion.

4. A few genres account for a large portion of the content, indicating that Netflix emphasizes popular categories.

5. Most titles were released in recent years, suggesting that Netflix prioritizes newer content.

Most Surprising Finding

One surprising finding was the rapid increase in content additions after 2015. Netflix expanded its content library significantly during this period. Another interesting observation was that movies greatly outnumber TV shows, highlighting Netflix's stronger emphasis on movies.
